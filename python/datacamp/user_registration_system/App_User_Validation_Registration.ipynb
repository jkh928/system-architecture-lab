```python
# Start coding here
# Use as many cells as you need
def validate_user(name, email, password):
    rules = [
        (validate_name(name), "name must be at least two characters long"),
        (validate_email(email), "email must be valid format (example@example.com"),
        (validate_password(password), "password must include capital letter, a number, and more than 8 characters")
    ]

    for is_valid, error_message in rules:
        if not is_valid:
            raise ValueError(error_message)
    return True

def register_user(name, email, password):
    try:
        validate_user(name, email, password)
        return {"name": name, "email": email, "password": password}
    except ValueError:
        return False

# Test scenario
test_users = [
    ("John Anatoli", "janatoli@example.com", "Password1234567890"),
    ("J", "johnnyappleseed@example.com", "Mypassword24")
]

for user_name, user_email, user_password in test_users:
    registration_result = register_user(user_name, user_email,user_password)

    if registration_result:
        message = f"User {user_name} registered successfully!"
        print(message)
        print(registration_result)
    else:
        print(f"\nUser {user_name} failed to register.")
    
```
